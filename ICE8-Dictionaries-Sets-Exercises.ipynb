### Exercise 1: Creating and Accessing a Dictionary
Create a dictionary called `capitals` that maps the following countries to their capital cities:
- France ‚Üí Paris
- Germany ‚Üí Berlin
- Italy ‚Üí Rome

Print the entire dictionary, and then print the capital of Germany using its key.

### Exercise 2: Modifying a Dictionary
Add a new country-capital pair to the `capitals` dictionary: Spain ‚Üí Madrid.
Update the capital of Italy to "Milan".
Delete the entry for France.
Print the dictionary after each operation.

````python
capitals = {
    "France": "Paris",
    "Germany": "Berlin",
    "Italy": "Rome"
} # Creating a dictionary of country capitals
capitals["Spain"] = "Madrid" # Adding a new country-capital pair
print("After adding Spain:", capitals)

capitals["Italy"] = "Milan" # Updating the capital of Italy
print("After updating Italy:", capitals)

del capitals["France"] # Deleting France from the dictionary
print("After deleting France:", capitals)
````

### Exercise 3: Dictionary Methods
Use the following dictionary:
```python
grades = {"Alice": 85, "Bob": 90, "Charlie": 78}
```
- Print all student names using `.keys()`
- Print all grades using `.values()`
- Print all student-grade pairs using `.items()`
- Use `.get()` to retrieve Diana‚Äôs grade with a default message.

````python
grades= {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78
} # Creating a dictionary of student grades
print("Student names:", (grades.keys())) #print student names
print("Grades:", (grades.values())) #print grades
print("Students and their Grades:", (grades.items())) #print students and their grades
print("Diana's grade:", grades.get("Diana", "Not Found")) #print Diana's grade with default value
````

### Exercise 4: Dictionary Comprehension
Create a list called `nums` with values `[1, 2, 3, 4, 5]`.
Use a dictionary comprehension to create a dictionary where the keys are the numbers and the values are their cubes.

````python
nums = [1, 2, 3, 4, 5]# Create a list of numbers
cubes = {n: n**3 for n in nums}# Create a dictionary with numbers and their cubes
print (cubes) #print the cubes dictionary 
````

### Exercise 5: Creating and Using Sets
Create a set named `colors` with the values: "red", "green", "blue", "green", "red".
Print the set. Add "yellow" to the set and then remove "blue".
Check if "purple" is in the set. Print the final set.

````python
colors = {"red", "green", "blue", "green", "red"} # Creating a set with some repeated colors
print("Original set:", colors) #print original set
colors.add("yellow") # Adding a new color
print("After adding yellow:", colors)
colors.remove("blue") # Removing blue color
print("After removing blue:", colors) 
print("Is 'purple' in the set?", "purple" in colors)# Check if purple is in the set 
print("Final set:", colors) #print final set
````

### Exercise 6: Set Operations
Create two sets:
```python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
```
Perform and print the result of the following operations:
- Union
- Intersection
- Difference (set1 - set2)
- Symmetric difference

````python
set1 = {1, 2, 3, 4} # First set
set2 = {3, 4, 5, 6} # Second set
print("Union:", set1.union(set2)) # Union - all unique elements from both sets
print("Intersection:", set1.intersection(set2)) #Intersection - common elements in both sets

print("Difference (set1 - set2):", set1.difference(set2))  #Difference - elements in set1 but not in set2

print("Symmetric Difference:", set1.symmetric_difference(set2)) #Symmetric Difference elements that are in one set OR the other, but not both
````
