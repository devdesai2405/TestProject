```python
%pip install pandas matplotlib

import pandas as pd
import matplotlib.pyplot as plt

# Optional: make plots look nicer
plt.style.use("default")
df = pd.read_csv(file_path)
df.head()
```

```python
# Convert date column to datetime (if not already)
df["date"] = pd.to_datetime(df["date"])

# Create extra time columns
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["hour"] = df["date"].dt.hour

df[["date", "year", "month", "hour"]].head()
```

```python
crimes_per_year = df.groupby("year").size()

print(crimes_per_year)

plt.figure(figsize=(8,5))
crimes_per_year.plot(marker="o")
plt.title("Total Crimes per Year (20152020)")
plt.xlabel("Year")
plt.ylabel("Number of Crimes")
plt.grid(True)
plt.show()
```

```python
top_types = df["primary_type"].value_counts().head(10)

plt.figure(figsize=(10,6))
top_types.sort_values().plot(kind="barh")
plt.title("Top 10 Crime Types (20152020)")
plt.xlabel("Number of Crimes")
plt.ylabel("Crime Type")
plt.tight_layout()
plt.show()
```

```python
crimes_by_hour = df.groupby("hour").size()

plt.figure(figsize=(10,5))
crimes_by_hour.plot(marker="o")
plt.title("Crimes by Hour of Day")
plt.xlabel("Hour (023)")
plt.ylabel("Number of Crimes")
plt.grid(True)
plt.show()
```

```python
crimes_by_month = df.groupby("month").size()

plt.figure(figsize=(10,5))
crimes_by_month.plot(marker="o")
plt.title("Crimes by Month")
plt.xlabel("Month (112)")
plt.ylabel("Number of Crimes")
plt.grid(True)
plt.show()
```

```python
arrest_rate = df["arrest"].value_counts(normalize=True) * 100

plt.figure(figsize=(6,5))
arrest_rate.plot(kind="bar")
plt.title("Overall Arrest vs Non-Arrest (%)")
plt.ylabel("Percentage")
plt.xticks(rotation=0)
plt.show()
```
