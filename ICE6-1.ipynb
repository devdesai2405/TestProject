## ICE 1: Word Counter

**Task:**
- Prompt the user to enter a sentence.
- Use the `split()` function to break the sentence into words.
- Print the total number of words.

```python
sentence = input("Enter a sentence: ") #asking the user for a sentence.
words = sentence.split() #now i am doing the split the sentence into word. 
print("total number of words:", len(words)) #printing the total number of words in the sentence.
```

## ICE 2: Case Conversion

**Task:**
- Define a string (for example, "Hello World").
- Print the string in lowercase using and in uppercase using appropirate methods.

```python
text = "Hello World" # defining a string 
print("Lowercase:", text.lower()) # so now i am converting to lowercase
print("Uppercase:", text.upper()) # same thing doing to uppercase
```

## ICE 3: Substring Extractor

**Task:**
- Given a string variable, extract the first 5 characters and the last 5 characters using slicing.
- If the string is shorter than 5 characters, print the string itself.

```python
text = "Programming"
if len(text) > 5: # so first doing the extract of first and last 5 characters
    print("String is shorter than 5 characters:", text)
else:
    print("First 5 characters", text[:5])
    print("Last 5 characters", text[-5:])
```

## ICE 4: Name List Splitter

**Task:**
- Define a string containing names separated by commas (e.g., "Alice, Bob, Charlie, Dana").
- Use `split()` method to convert the string into a list of names.
- Print the resulting list.

```python
names = "Alice, Bob, Charlie, Dana" 
name_list = names.split(",") # doing the split the string into a list using commas 
name_list = [name.strip() for name in name_list] #removing the extra space around each name
print("List of names:", name_list) # printing the resulting list
```

## ICE 5: Character Replacement

**Task:**
- Prompt the user for a sentence.
- Replace every occurrence of the letter `'a'` with `'@'` using the appropriate method.
- Print the modified sentence.

```python
sentence = input("Enter a sentence: ")
modified_sentence = sentence.replace("a", "@") #replacing every 'a' with '@'
print("Modified sentence", modified_sentence) #printing the modified sentence 
```
